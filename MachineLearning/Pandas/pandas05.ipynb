- qcut() : 범주를 만든다(동일한 갯수) - 그룹에 대한 통계량을 확인 할 수 있다.  
- cut()  : 범주를 만든다(동일한 길이) - 그룹에 대한 통계량을 확인 할 수 있다.

```python
print('각 붓꽃 꽃잎의길이가 해당 종내에서 대/중/소 어느그룹에 해당되는지 알고싶다면? - ')
# display( iris_datasets )
iris_grp = iris_datasets.groupby('species').petal_length
for key , group in iris_grp :
    print('key - ' , key)
    print('len - ' , len(group))
    print()
    print(group)
    print("*" * 50)
```

```python
iris_datasets.groupby('species').petal_length.size()
```

```python
def cut_petal_length(s) :
    return pd.qcut(s  , 3 , labels=['대', '중', '소'])
```

- transform() : 데이터프레임 자체를 변환시켜주는 함수

```python
iris_datasets['category'] = iris_datasets.groupby('species').petal_length.transform(cut_petal_length)
# iris_datasets['category'] = iris_datasets.groupby('species').petal_length.apply(cut_petal_length)
# iris_datasets['category'] = iris_datasets.groupby('species').petal_length.agg(cut_petal_length)
```

```python
print('종별로 꽃잎길이 , 꽃잎폭 평균을 구한다면? - ')
print('agg - ')
iris_datasets.groupby('species').agg(np.mean).loc[ :  , ['petal_length' , 'petal_width']]
```

```python
print('종별로 최대/최소값을 구한다면? - ')
print('agg - ')
iris_datasets.groupby('species').agg([np.min, np.max])
```

```python
print('[문제01] - ')
print('qcut 명령으로 세 개의 나이 그룹을 만든다. - ')
titanic_datasets = sns.load_dataset('titanic')
titanic_datasets['age_category'] = pd.qcut(titanic_datasets['age'] , 3 , labels = ['청소년' , '성인' , '노인' ])
titanic_datasets
```

```python
print()
print('[문제02] - ')
print('성별, 선실, 나이그룹(age_category) 의한 생존율을 데이터프레임으로 생성하고')
print('행에는 성별 및 나이그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다 -')
print('생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다. -')
print('cut - pivot_table')

frm = titanic_datasets.pivot_table(index=['sex','age_category'] , 
                             columns = 'pclass' , values = 'survived', 
                             margins = True , 
                             margins_name = 'survived_ratio' , 
                             aggfunc = 'count')
display( frm )
print()
# print( type( frm.loc['survived_ratio' , 'survived_ratio'] ))
survived_frm = frm / frm.loc['survived_ratio' , 'survived_ratio'][0]
display( survived_frm )
```

```python
print()
print('문제[03] - ')
print('성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다. - ')

frm = titanic_datasets.pivot_table(index='sex' , 
                             columns = 'pclass' , values = 'survived', 
                             margins = True , 
                             margins_name = 'survived_ratio' , 
                             aggfunc = 'count')
display(frm)

print()
survived_frm = frm / titanic_datasets.shape[0]
display(survived_frm)
```
