- 서울지역 대학교 위치.xlsx

```python
location_frm = pd.read_excel('./data/서울지역 대학교 위치.xlsx')
location_frm.set_index('Unnamed: 0' , inplace=True)
location_frm.shape[0]
```

```python
map = g.Map(location=[37.5532035880752, 126.98739716743572] , 
            zoom_start=12 , 
            tiles = 'Stamen Terrain')
```

```python
print('대학교 위치정보를 Marker로 표시')
print()
# g.Marker([lat, lng] , popup=name).add_to(map))

for name, lat, lng in zip(location_frm.index , location_frm.위도 , location_frm['경도']) :
    # print(name , lat , lng)
    g.Marker([lat, lng] , popup=name , icon = g.Icon(color='red')).add_to(map)

map
```

```python
print('대학교 위치정보를 CircleMarker로 표시')
print()
# g.Marker([lat, lng] , popup=name).add_to(map))

for name, lat, lng in zip(location_frm.index , location_frm.위도 , location_frm['경도']) :
    # print(name , lat , lng)
    g.CircleMarker([lat, lng] , 
                    popup=name , 
                    icon = g.Icon(color='blue') , 
                    fill = True , 
                    fill_color = 'balck' , 
                    fill_opacity = .5 , 
                    radius = 10).add_to(map)

map
```

- elem_school_folium.xlsx

```python
school_frm = pd.read_excel('./data/elem_school_folium.xlsx')
school_frm.head()
```

```python
print('Feature Selection - 학교명, 학교급구분, 소재지도로명주소, 위도, 경도')

subset_frm = school_frm.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
subset_frm.head()
```

```python
print('조건 - 소재지도로명주소에 서울이 포함되고 학교급구분이 초등학교만 추출한다면? ')
print('hint - str.contains()')
print()
```

```python
print('case01 - ')
print('결측값 확인 - ')
print( subset_frm.isnull().sum() )
print()
print('결측값 drop - ')
subset_frm.dropna(inplace = True)
print( subset_frm.isnull().sum() )

condition = "학교급구분 == '초등학교' and 소재지도로명주소.str.contains('서울특별시').values"
filter_frm = subset_frm.query( condition  )
filter_frm
```

```python
seoul_school_map = g.Map(   location=[37.5532035880752, 126.98739716743572] , 
                            zoom_start=12 , 
                            tiles = 'Stamen Terrain')

for idx in filter_frm.index :
    g.Marker([ filter_frm.loc[idx, '위도'] , filter_frm.loc[idx, '경도'] ] , 
             popup =filter_frm.loc[idx, '학교명']).add_to(seoul_school_map)
        
seoul_school_map
```

- Choropleth

```python
pop_frm = pd.read_excel('./data/경기도인구데이터.xlsx')
pop_frm.set_index('구분' , inplace=True)
pop_frm.head()
```

```python
geo_data = json.load( open('./data/경기도행정구역경계.json' , encoding='utf-8') )
geo_data
```

```python
pop_frm.columns = pop_frm.columns.map(str)
pop_frm.columns 
```

```python
gyeonggi_map = g.Map(location=[37.26385623844609, 127.02878746071707] , 
                            zoom_start=12, 
                            tiles = 'Stamen Terrain')

print('구분된 경계선을 표시하는 객체 - ')
print()
print()
print()

g.Choropleth(geo_data = geo_data , 
             data     = pop_frm['2017'] , 
             columns  = [pop_frm.index , pop_frm['2017']] ,
             key_on   = 'feature.properties.name').add_to(gyeonggi_map)

gyeonggi_map
```
