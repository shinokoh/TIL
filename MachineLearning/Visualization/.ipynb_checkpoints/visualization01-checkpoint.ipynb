```python
datasets = pd.read_excel('./data/시도별 전출입 인구수.xlsx' , encoding = 'utf-8')
#datasets = pd.read_excel('./data/시도별 전출입 인구수.xlsx')
datasets
```

```python
# nan  을 채워넣기위해 fillna
# fillna(대체값 | 'missing')
# 여기서 missing은 fill gaps forward or backword 을 ffill 이라고 줄 수 있다. backword는 bfill ?
```

```python
datasets = datasets.fillna(method='ffill')
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 서브셋을 만든다면?

seoul_frm = datasets[(datasets['전출지별']=='서울특별시') & (datasets['전입지별']!='서울특별시')]
```

```python
# 전출지별을 삭제?
seoul_frm = seoul_frm.drop(['전출지별'] ,axis=1)
seoul_frm
```

```python
# 전입지별 인덱스 이름을 전입지로 변경?

# 컬럼명 변경 딕셔너리 이용 바로 적용시키려면 inplace True
seoul_frm.rename({'전입지별' : '전입지'} , axis = 1 , inplace = True)

#전입지를 index로 만들려면
seoul_frm.set_index('전입지', inplace=True)
```

```python

# 경기도 이동한 인구 데이터 값만 선택해서 시각화
plt.plot( seoul_frm.columns , seoul_frm.loc['경기도'] )
plt.title("서울에서 경기")
plt.xticks(rotation = 'vertical') # x 축값들 최소한으로 겹치게
plt.show()

seoul_frm
```
