- 시도별 전출입 인구수.xlsx 를 이용한 라인플롯
- fillna(대체값 | 'missing' | (fill gaps forward or backword) ffill bfill)

```python
dataset = pd.read_excel('./data/시도별 전출입 인구수.xlsx' , header = 0 )
```

```python
dataset.head()
```

```python
dataset = dataset.fillna(method='ffill')
dataset.head()
```

```python
print('서울에서 다른 지역으로 이동한 데이터만 추출하여 서브셋을 만든다면? - ') 

mask = (dataset['전출지별'] == '서울특별시') & (dataset['전입지별'] != '서울특별시')
seoul_frm = dataset[mask]
seoul_frm.head()
```

```python
print('전출지별을 삭제한다면? - ')
seoul_frm = seoul_frm.drop(['전출지별'] , axis = 1)
seoul_frm.head()
```

```python
print('전입지별 인덱스 이름을 전입지로 변경한다면? - ')
print('rename - ')

print()
seoul_frm.rename({'전입지별' : '전입지'} , axis = 1 , inplace = True )
seoul_frm.head()
```

```python
print('전입지를  인덱스로 변경한다면? - ')
print('set_index - ')
print()

seoul_frm.set_index('전입지' , inplace = True)
seoul_frm.head()
```

```python
datas = seoul_frm.loc['경기도']
print('type  - ' , type(datas))
print()
print('index - ' , datas.index)
print()
print('values - ' , datas.values)
print()
```

```python
print('경기도 이동한 인구 데이터 값만 선택해서 시각화?')
print()
print()

plt.figure(figsize = (15, 5))

plt.plot(datas.index , datas.values , marker='+' , markersize = 4)

plt.title('서울 -> 경기 인구이동 수 ')
plt.xlabel('년도')
plt.ylabel('이동인구 수')
plt.xticks(rotation = 'vertical')

plt.legend(labels = ['서울 -> 경기'] , loc = 'best')

plt.ylim(5000 , 700000)

plt.show()
plt.close()
```

```python
print('subplot 이용한 시각화?')
print()
print()

fig = plt.figure(figsize = (20, 10))

ax01 = fig.add_subplot(1, 2, 1)
ax02 = fig.add_subplot(1, 2, 2)


ax01.plot(datas.index , datas.values , marker='+' , markersize = 4)
ax01.set_xlabel('년도')
ax01.set_ylabel('이동인구 수')
ax01.set_xticklabels(labels = datas.index  , rotation = 75)
ax01.legend(labels = ['서울 -> 경기'] , loc = 'best')
ax01.set_ylim(5000 , 700000)



ax02.plot(datas.index , datas.values , marker='+' , markersize = 4)
ax02.set_xlabel('년도')
ax02.set_ylabel('이동인구 수')
ax02.set_xticklabels(labels = datas.index  , rotation = 75)
ax02.legend(labels = ['서울 -> 경기'] , loc = 'best')
ax02.set_ylim(5000 , 700000)




plt.show()
plt.close()
```

- bar plot (plt.bar(X , Y))
- X -> 범주형데이터

```python
import seaborn as sns
iris_datasets = sns.load_dataset('iris')
titanic_datasets = sns.load_dataset('titanic')
```

```python
print('index  - ' , titanic_datasets.groupby('pclass').survived.sum().index )
print('values - ' , titanic_datasets.groupby('pclass').survived.sum().values )
```

```python
print('선실등급에 따른 생존자 수를 시각화 한다면?  - ') 
print()
print()

X =  titanic_datasets.groupby('pclass').survived.sum().index 
Y =  titanic_datasets.groupby('pclass').survived.sum().values 

plt.figure(figsize = (15, 5))

plt.bar(X, Y)

plt.title('선실등급에 따른 생존자 수 ')
plt.xlabel('선실등급')
plt.ylabel('생존자 수')
plt.xticks(X)
plt.legend(labels = ['생존자 수'] , loc = 'best')


plt.show()
plt.close()
```

```python
print('선실등급에 따른 생존자 수를 시각화 한다면?  - ') 
print()
print()

X =  titanic_datasets.groupby('pclass').survived.sum().index 
Y =  titanic_datasets.groupby('pclass').survived.sum().values 

plt.figure(figsize = (15, 5))

plt.barh(X, Y)

plt.title('선실등급에 따른 생존자 수 ')
plt.xlabel('생존자 수')
plt.ylabel('선실등급')
plt.xticks(Y)
plt.yticks(X)
plt.legend(labels = ['생존자 수'] , loc = 'best')


plt.show()
plt.close()
```
